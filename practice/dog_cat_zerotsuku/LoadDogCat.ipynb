```python
import matplotlib.pyplot as plt
import os
import cv2
import random
import numpy as np
```

```python
DATADIR = "./dataset/kagglecatsanddogs_5340/PetImages/"
CATEGORIES = ["Dog", "Cat", "Dog_test", "Cat_test"]
IMG_SIZE = 80
training_data = []
```

```python
for class_num, category in enumerate(CATEGORIES):
    print("{0}:{1}".format(class_num, category))
```

```python
def create_training_data():

    for class_num, category in enumerate(CATEGORIES):
        path = os.path.join(DATADIR, category)
        for image_name in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, image_name),)
                #image reading
                img_resize_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                #resize the image
                training_data.append([img_resize_array, class_num])
                #append image data, label info

                
            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
random.shuffle(training_data)
```

```python
X_train = []
Y_train = []
```

```python
for feature, label in training_data:
    X_train.append(feature)
    Y_train.append(label)
```

```python
X_train = np.array(X_train)
Y_train = np.array(Y_train)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
for i in range(0, 4):
    print("label of training data", Y_train[i])
    plt.subplot(2, 2, i+1)
    plt.axis('off')
    plt.title('Dog' if Y_train[i] == 0 
              else 'Cat' if Y_train[i] == 1
              else 'Dog_test' if Y_train[i] == 2
              else 'Cat_test')
    plt.imshow(X_train[i], cmap='gray')
plt.show()
```
